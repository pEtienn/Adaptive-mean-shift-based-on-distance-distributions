```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as sStats
from sklearn.datasets import make_blobs
import sklearn.metrics as metrics
import fctMS
```

# Generate Dataset

### 4 2D clusters with varying parameters

```python
means=np.array([[0,0],[4.2,4.2],[-5,5],[2.7,-2.5]])
scales=np.array([1,1.5,2,0.7])
ns=np.array([50,75,200,25])

[X,labels]=fctMS.generate2DClusters(ns,means,scales)
```

### 3 similar blobs

```python

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels = make_blobs(n_samples=1000, centers=centers, cluster_std=0.6)
```

# Adaptive Mean shift

```python
ms=fctMS.AdaptiveMeanShift(X,minClusterSize=10,maxClusterSize=0.7)
ms.meanShift()
fctMS.print2DClusters(ms.X,ms.labels,ms.clusterCentroids)
fctMS.print2DNestimates(ms._XW.X,ms._estimatedClusterSize)
```

## Test the result

```python
metrics.homogeneity_score(labels,ms.labels)
```
